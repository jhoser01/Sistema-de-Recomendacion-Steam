```python
#cargamos las librerías
import gzip
import json
import ast
import pandas as pd
```

<h1>1. Cargar los datos</h1>

<h3>Observamos cómo está estructurado el archivo</h3>

```python
archivo_gz = 'G:/Mi unidad/BOOTCAMPS/SOYHENRY/CIENCIA DE DATOS/CURSO/PT07/PI_MLOPS/DATASET_ORIGINAL/steam_games.json.gz'
with gzip.open(archivo_gz, 'rb') as f:
    for _ in range(10):
        linea = f.readline()  # Lee una línea del archivo
        if not linea:  # Si no hay más líneas en el archivo, sal del bucle
            break
        json_bytes = linea.decode('utf-8')  # Decodifica la línea en una cadena de texto
        print(json_bytes)  # Imprime la línea
```

```python
def converter(ruta):
    data = []
    with gzip.open(ruta, 'rt', encoding='utf-8') as file:
        for line in file:
            try:
                obj = json.loads(line.strip())
                data.append(obj)
            except json.JSONDecodeError as e:
                print(f"Error al decodificar JSON en línea: {line.strip()}")
                print(str(e))
    return data    
datos = converter(archivo_gz)

dfgames = pd.DataFrame(datos)
dfgames
```

<h3>Lo convertimos a csv</h3>

```python
dfgames.to_csv("G:/Mi unidad/BOOTCAMPS/SOYHENRY/CIENCIA DE DATOS/CURSO/PT07/PI_MLOPS/DATASET_ORIGINAL/CSV/games.csv", index=False)
```
