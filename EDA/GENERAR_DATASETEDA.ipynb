<h1>Explotamos el dataframe de item y review</h1>

```python

# Función para explotar dos columnas simultáneamente
def explode_columns(df, col1, col2, extra_col):
    # Combinar las listas de ambas columnas utilizando zip y explotarlas
    exploded = df.apply(lambda row: list(zip(row[col1], row[col2], [row[extra_col]] * len(row[col1]))), axis=1).explode()
    # Convertir la serie resultante en un DataFrame
    exploded_df = exploded.apply(pd.Series)
    # Asignar nombres de columnas
    exploded_df.columns = [col1, col2, extra_col]
    return exploded_df.reset_index(drop=True)
```

```python
df_item = explode_columns(item, 'item_id', 'playtime_forever', 'user_id')
df_item.head(30)
```

```python
# Función para explotar dos columnas simultáneamente
def explode_columns2(df, col1, col2, col3, col4, extra_col):
    # Combinar las listas de ambas columnas utilizando zip y explotarlas
    exploded = df.apply(lambda row: list(zip(row[col1], row[col2], row[col3], row[col4],[row[extra_col]] * len(row[col1]))), axis=1).explode()
    # Convertir la serie resultante en un DataFrame
    exploded_df = exploded.apply(pd.Series)
    # Asignar nombres de columnas
    exploded_df.columns = [col1, col2, col3, col4, extra_col]
    return exploded_df.reset_index(drop=True)
```

```python
df_review = explode_columns2(review, 'item_id', 'recommend', 'year_review', 'sentiment_analysis', 'user_id')
df_review.head(30)
```

```python
merged_df = pd.merge(df_item, df_review, on=['user_id', 'item_id'], how='inner')
```
